```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_path = '../src/main/resources/input/data_scenarios_a_example.in'
input_path = '../src/main/resources/input/data_scenarios_b_mumbai.in'
input_path = '../src/main/resources/input/data_scenarios_c_metropolis.in'
input_path = '../src/main/resources/input/data_scenarios_d_polynesia.in'
input_path = '../src/main/resources/input/data_scenarios_e_sanfrancisco.in'
input_path = '../src/main/resources/input/data_scenarios_f_tokyo.in'

with open(input_path, 'r') as file:
    lines = file.readlines()

W, H = [int(i) for i in lines[0].split()]
N, M, R = [int(i) for i in lines[1].split()]
building = np.zeros(shape=(N,4), dtype=np.int32)
antenna = np.zeros(shape=(M,2), dtype=np.int32)
startRow = 2
for i in range(N):
    building[i] = [int(i) for i in lines[startRow+i].split()]
startRow += i+1
for i in range(M):
    antenna[i] = [int(i) for i in lines[startRow+i].split()]

# plot
bc = np.zeros(shape=(W,H), dtype=np.float32)
for b in building:
    bc[b[0],b[1]] = b[3]

plt.figure(figsize=(10,10))
plt.imshow(bc)
plt.show()
```
